```python
from math import sqrt, sin
f = lambda x: x**4 + sin(23*x)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-1.5, 1.5, 100000)
y = map(f, x)

plt.plot(x, y, label = "f(x)", color = 'g')
plt.legend(['f(x)'], loc='upper left')

plt.show()
```

## To find the global minimum, we can suppose that x is in (-1.0, +1.0)

```python
phi = (1 + sqrt(5))/2
resphi = 2 - phi
 
# a and b are the current bounds; the minimum is between them.
# c is the center pointer pushed slightly left towards a
def goldenSectionSearch(f, a, c, b, absolutePrecision):
    if abs(a - b) < absolutePrecision:
        return (a + b)/2
    # Create a new possible center, in the area between c and b, pushed against c
    d = c + resphi*(b - c)
    if f(d) < f(c):
        return goldenSectionSearch(f, c, d, b, absolutePrecision)
    else:
        return goldenSectionSearch(f, d, c, a, absolutePrecision)
```

### Golden Section Search in (-1, 1)

```python
print(' x                     f(x)')
minimum_x = goldenSectionSearch(f, -1., -1. + 2*resphi, 1., 1e-10)
print(minimum_x, f(minimum_x))
```

### Golden Section Search in (-0.5, 0.5)

```python
print(' x                     f(x)')
minimum_x = goldenSectionSearch(f, -0.5, -0.5 + 1*resphi, 0.5, 1e-10)
print(minimum_x, f(minimum_x))
```

### Two golden section search results differs. The latter is the global optimal.

```python
# a and b are the current bounds; the minimum is between them.
# c is the center pointer pushed slightly left towards a
def Brent_Method(f, a, c, b, absolutePrecision):
    if abs(a - b) < absolutePrecision:
        return (a + b)/2
    # Create a new possible center, in the area between c and b, pushed against c
    numerator = ((b-a)**2)*(f(b)-f(c)) - ((b-c)**2)*(f(b)-f(a))
    denominator = 2*((b-a)*(f(b)-f(c)) - (b-c)*(f(b)-f(a)))
    d = b - numerator/denominator
    if f(d) < f(c):
        return Brent_Method(f, c, d, b, absolutePrecision)
    else:
        return Brent_Method(f, d, c, a, absolutePrecision)
```

### Brent's Method in (-1, 1)

```python
print(' x                     f(x)')
minimum_x = Brent_Method(f, -1., (-1.+1.)/2, 1., 1e-10)
print(minimum_x, f(minimum_x))
```

### Brent's Method in (-0.5, 0.5)

```python
print(' x                     f(x)')
minimum_x = Brent_Method(f, -0.5, 0, 0.5, 1e-10)
print(minimum_x, f(minimum_x))
```
